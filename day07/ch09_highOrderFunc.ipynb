## Chapter09. 함수
### section 04) 고차 함수

```python
"""함수가 객체인지 확인"""

def min2(a, b):
    """a, b의 최솟값을 구해 리턴하는 함수"""
    return a if a < b else b

a = int(input('정수 a：'))
b = int(input('정수 b：'))

func = min2 # 함수는 객체 - 참조변수가 가리킬 수 있음
print('최솟값은', func(a, b), '입니다.') # 참조변수가 함수 객체를 가리키면 함수 호출 가능

del min2
print('최솟값은', func(a, b), '입니다.')

del func
print('최솟값은', func(a, b), '입니다.')
```

```python
"""하나의 이름으로 두 함수 호출하기"""

def mul2(x, y):
    return x * y

def add2(x, y):
    return x + y

a = int(input('정수 a：'))
b = int(input('정수 b：'))

func = mul2
print('a와 b의 곱은', func(a, b), '입니다.')

func = add2
print('a와 b의 합은', func(a, b), '입니다.')
```

```python
"""9행 9열의 곱셈과 덧셈표 출력하기"""

def gugu(func): # 고차원 함수
    """9행 9열의 표를 출력"""
    for i in range(1, 10):
        for j in range(1, 10):
            print('{:3d}'.format(func(i, j)), end='')
        print()

def mul2(x, y):
    return x * y

def add2(x, y):
    return x + y

n = int(input('곱셈[0]/덧셈[1]：'))

if n == 0:
    print('9행 9열의 곱셈 표')
    gugu(mul2)
elif n == 1:
    print('9행 9열의 덧셈 표')
    gugu(add2)
```
