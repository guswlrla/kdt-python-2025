```python
"""두 수의 합을 구하는 람다식"""

a = int(input('정수 a：'))
b = int(input('정수 b：'))

add2 = lambda x, y: x + y
print('a와 b의 합은', add2(a, b), '입니다.')
```

```python
"""9행 9열의 곱셈과 덧셈표 출력하기"""

def gugu(func):
    """9행 9열의 표를 출력"""
    for i in range(1, 10):
        for j in range(1, 10):
            print('{:3d}'.format(func(i, j)), end='')
        print()

n = int(input('곱셈[0]／덧셈[1]：'))

if n == 0:
    print('9행 9열의 곱셈 표')
    gugu(lambda x, y: x * y)
elif n == 1:
    print('9행 9열의 덧셈 표')
    gugu(lambda x, y: x + y)
```

```python
# 리스트의 모든 요소를 두배로 만들기(함수)

def double(n):
    return 2 * n

x = [1, 2, 3, 4]
y = map(double, x)

print(list(y))
```

```python
# 리스트의 모든 요소를 두배로 만들기(람다식)

x = [1, 2, 3, 4]
y = map(lambda n: 2 * n, x)

print(list(y))
```

```python
# 80점 이상의 점수만 추출하기

import random

number = int(input('학생 수 : '))

score = [None] * number

for i in range(number):
    score[i] = random.randint(0, 100)

print('모든 학생 점수＝', score)
print('합격자 점수', list(filter(lambda n: n >= 80, score)))
```
