```python
# 어노테이션이 있는 함수 puts

def puts(n: int, s: str) -> None: # 표준 코딩 방식 - 콜론(:) 뒤에 공백, 화살표(->) 앞뒤에 공백
    """n개의 s를 연속해서 출력"""
    for _ in range(n):
        print(s, end='')

puts(5, '*')
print()
print(puts.__annotations__) # 함수의 어노테이션을 딕셔너리화
print()
# puts('*', 5) # 순서가 바뀌어서 오류
```

```python
"""어노테이션과 문서화 문자열이 있는 함수 puts"""


def puts(n: int, s: str) -> None:
    """n개의 s를 연속해서 출력

    매개 변수:
        n -- 출력하는 문자열의 개수
        s -- 출력하는 문자열
    리턴값:
        없음

    """
    for _ in range(n):
        print(s, end='')

# 비교
print(puts.__doc__) # 문서화 문자열을 출력
help(puts)
print(puts.__annotations__)
```

```python
"""평균값 구하기"""

def ave(*args) -> float: # *ㅇㅣ 뭐지
    """가변 인수의 평균 구하기"""
    return sum(args) / len(args)

print('ave(1, 2, 3) = {}'.format(ave(1, 2, 3)))
print('ave(5, 7.77, 5) = {}'.format(ave(5, 7.77, 5)))
print('ave(3.5, 4.7, 8.2) = {}'.format(ave(3.5, 4.7, 8.2)))
```

```python
"""리스트 형식의 문자열로 변환하기"""

def list_str(*args) -> str:
    """가변 인수를 리스트 형식의 문자열으로
       변환하여 리턴"""
    return str(list(args))

print('list_str(1, 2, 3) = {}'.format(list_str(1, 2, 3)))
print('list_str(5, 7.77, 5) = {}'.format(list_str(5, 7.77, 5)))
print('list_str(3.5, 4.7, 8.2) = {}'.format(list_str(3.5, 4.7, 8.2)))
```
