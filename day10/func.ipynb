```python
def add(a: int, b: int) -> int:
    if isinstance(a, int) and isinstance(b, int): # 방어코드 - 정수일 경우에만 처리
        return a+b
    raise Exception("정수가 아닙니다.")

add(3, 4)
```

```python
def is_leapyear(year: int) -> bool:
    return year % 4 == 0 and year % 100 != 0 or year % 400 == 0

is_leapyear(2025)
```
