```python
# 집합 생성

# 방법1: 중괄호로 생성
s1 = {1, 2, 3}
print(s1)
print(type(s1))

# 방법2: set() 함수로 생성
s2 = set('ABC')
print(s2)
s3 = set([1, 2, 3, 4])
print(s3)


s3.add(5)
print(s3)

## 1과 5가 같은 집합인지 알아내는법
n = 1
m = 5

if n in s3 and m in s3:
    print('사이클')
```
