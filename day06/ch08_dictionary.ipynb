### dictionary
- 모든 요소가 키(key)와 값(value)의 쌍
- 단어(key) : 뜻(value)
- json이 대표적인 예

```python
season = {
    '봄': 'spring',
    '여름': 'summer',
    '가을': 'autumn',
    '겨울': 'winter',
}

print("season['봄'] = ", season['봄'])
# print("season['늦봄'] = ", season['늦봄']) # 존재하지 않는 키로 접근하면 오류
```

```python
# 딕셔너리 생성

# 방법1 : 중괄호로 생성
d1 = {'Korea' : 410, 'China' : 156, 'Japan' : 392, 'Taiwan' : 158}
print(d1)

# 방법2 : dict() 함수로 생성
lst = [('Korea', 410), ('China', 156), ('Japan', 392), ('Taiwan', 158)] # 튜플 리스트 이용
d2 = dict(lst)
print(d2)

key = ['Korea', 'China', 'Japan', 'Taiwan']
val = [410, 156, 392, 158]
d3 = dict(zip(key, val)) # zip() 이용
print(d3)
```

```python
# 요소의 순서와 등가 판정
lst1 = ['Korea', 'China', 'Japan']
lst2 = ['Korea', 'Japan', 'China']
print(lst1 == lst2) # 리스트는 요소의 값이 같아도 순서가 다르면 등가 성립x

dic1 = {'Kor' : 'Korea', 'CHN' : 'China', 'JPN' : 'Japan'}
dic2 = {'Kor' : 'Korea', 'JPN' : 'Japan', 'CHN' : 'China'}
print(dic1 == dic2) # 딕셔너리는 요소(키와 값의 쌍)가 같다면 순서가 달라도 동일
print(dic1['Kor']) # 인덱싱 할 때, 키 값을 사용
```

```python
# 인덱스식과 get 메소드
color = {'red' : '빨강', 'green' : '초록', 'blue' : '파랑'}
# print(color['pink']) # 없는 키로 접근하면 오류
print(color.get('pink')) # get() - KeyError 예외 발생을 막고, 값을 추출함
```

```python
# 인덱스식과 setdefault 메소드
color = {'red' : '빨강', 'green' : '초록', 'blue' : '파랑'}
print(color)

color['red'] = '빨간색' # 키 값이 존재하면 업데이트 
print(color)

color['yellow'] = '노랑' # 키 값이 없으면 삽입
print(color)

# setdefault()
print(color.setdefault('yellow', '노랑')) # 키 값이 존재하면 값 리턴, 없으면 삽입(value를 생략하면 None)
color.setdefault('pink', '분홍')
print(color)
```

```python
# update 메소드로 딕셔너리 수정
rgb = {'red': '빨강', 'green': '초록', 'blue': '파랑'}
cmy = {'cyan': '하늘', 'magenta': '자주', 'yellow': '노랑'}

rgb.update(cmy) # rgb에 cmy 결합
print(rgb)
rgb.update(red='레드', blue='블루', yellow='옐로우') # 딕셔너리 요소 수정(키가 있으면 수정, 없으면 삽입)
### 함수의 매개변수 전달 방식 : 키워드 전달 방식
print(rgb)
```

```python
# 요소 제거

# 방법1: pop()으로 제거
rgb.pop('yellow') # 키 값이 없으면 에러
print(rgb)

# 방법2: del문으로 제거
del rgb['magenta'] # 키 값이 없으면 에러
print(rgb)

# 방법3: clear 메소드로 전체 요소 제거
rgb.clear()
print(rgb)
```

```python
# 딕셔너리 기본 조작
if 'purple' not in color: # in, not in 연산자 이용
    color['purple'] = '보라'
print(color)
```

```python
# 딕셔너리 탐색
rgb = {'red': '빨강', 'green': '초록', 'blue': '파랑'}

for i, j in enumerate(rgb): # 딕셔너리는 순회하면 키 값이 나와서 키를 출력
    print('{} {}'.format(i, j))

for i, key in enumerate(rgb, 1): # 카운트 1부터 시작
    print('{} {}'.format(i, key))

for key in rgb: # 인덱스 사용x
    print('{}'.format(key))
```

```python
# 딕셔너리와 뷰
rgb = {'red': '빨강', 'green': '초록', 'blue': '파랑'}

print('키：', list(rgb.keys()))       # 키의 뷰를 리스트로 변환
print('값：', list(rgb.values()))    # 값의 뷰를 리스트로 변환
print('요소：', list(rgb.items()))     # (키, 값)의 뷰를 리스트로 변환
```

```python
# 딕셔너리 내포표기
txt = 'abcabcde'
count = {ch : txt.count(ch) for ch in txt} # {키 : 값 for 요소 in 이터러블}
print(count)

s = {0 : '봄', 1 : '여름', 2 : '가을', 3 : '겨울'}
seasons = {k : v for k, v in s.items()} # enumerate()는 키만 순회함
print(seasons)
```
