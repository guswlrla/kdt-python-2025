```python
# 직각 이등변 삼각형과 사각형

def put_star(n):
    """n개의 '*'을 연속해서 출력"""
    for _ in range(n):
        print('*', end='')

print('직각 이등변 삼각형')
n = int(input('짧은 변의 길이：'))

for i in range(1, n + 1):
    put_star(i)
    print()

print('사각형')
w = int(input('가로：'))
h = int(input('세로：'))

for i in range(1, h + 1):
    put_star(w)
    print()
```

```python
# 두 값의 최솟값과 최댓값 구하기

def min_max2(a, b):
    """a와 b의 최솟값과 최댓값을 구하여 리턴"""
    return (a, b) if a < b else (b, a)

n1 = int(input('정수 n1：'))
n2 = int(input('정수 n2：'))

minimum, maximum = min_max2(n1, n2) # pack한 뒤 unpack
print('최솟값은', minimum, '이고 최댓값은', maximum, '입니다.')
```

```python
# 1부터 n까지의 합 구하기

def sum_1ton(n):
    """1부터 n까지의 합 구하기"""
    s = 0
    while n > 0:
        s += n
        n -= 1
    return s

x = 5
total = sum_1ton(x)
print('1부터 ', x, '까지의 합은 ', total, '입니다.', sep='')
```

```python
# 리스트의 요소 값 변경하기

def change(lst, idx, val):
    """lst[idx]의 값을 val으로 변경"""
    lst[idx] = val
    print(lst)

x = [11, 22, 33, 44, 55]
y=x[:]
print(id(x), id(y))
x2=x
print(id(x), id(x2))
index = 2
value = 99

# change(x, index, value)
change(x[:], index, value)
print('x =', x)
```

```python
# 이름에 호칭을 붙여 인사하는 함수(인수의 기본값 사용하기)

def hello(name, honorific = '씨'):
    """호칭을 붙여서 인사하기"""
    print('안녕하세요, {} {}.'.format(name, honorific))

hello('박시은')
hello('김지연', '선생님')
hello('이서현', '님')
hello('정승제', '생선님')
```
