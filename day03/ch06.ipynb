### enumerate 함수

```python
s = 'hello'

for idx, i in enumerate(s):
    print(idx, i)
```

```python
a = 'hello'
b = [1, 2, 3, 4 , 5]

for i, j in zip(a, b): # zip()으로도 가능
    print(j, i)
```
