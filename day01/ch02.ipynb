### input 함수를 사용해서 문자열 입력받기

```python
# 1 2 3 4 5
data = input()
print(type(data)) # input 함수는 항상 문자열로 입력 받음, '1 2 3 4 5'
```

```python
# 입력받은 문자열을 더하기
data = input()

# 1단계. split 함수를 사용해서 공백을 기준으로 분리
data = data.split(" ") # split 디폴트 값은 공백
print(data)

# 2단계. 정수로 변환해서 리스트에 저장
data = list(map(int, data))
print(data)

# 3단계. 정수 합계
print(sum(data))
```

### format 메소드를 이용해서 문자열 서식화

```python
a = 5
print('result : {}'.format(a)) # 'result : 5'
b = 3.141592
print(f'result : {b:.2f}') # .2f - 소수점 둘째 자리까지 표시
```
