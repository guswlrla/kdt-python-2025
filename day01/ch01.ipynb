### 연산자와 피연산자
- 연산을 위해 사용하는 기호는 연산자(operator), 연산의 대상을 피연산자(operand)라고 함

```python
print(7 + 3)
print(7 - 3)
print(7 * 3)
print(7 / 3) # 자바나 c에선 정수 값이 나오지만, 파이썬은 실수가 나옴
print(7 // 3) # 몫
print(7 ** 3) # 거듭제곱, Math.pow(7, 3)
```

### 수치 자료형
- 수나 문자를 나타내는 종류나 방법을 자료형(data type)
- 파이썬에는 수치 자료형 3개가 있음(`int`, `float`, `complex`(복소수))

```python
lst = [1, 3.14, 'hello', True] # 파이썬은 모든 것이 object타입
lst
```

```python
for l in lst:
    print(type(l), l)
```

### 숫자 리터럴
- 수의 값을 나타내는 방법
- ex) 7은 정수 리터럴(integer literal), 3.0은 실수 리터럴(floating point literal)

```python
a = 100_000_000 # 파이썬은 숫자 중간에 있는 밑줄을 무시(큰 값을 읽기 쉽게 할 때 쓰임)
a
```

```python
# e는 n.0에서의 점을 나타냄
# e 뒤에 양수 -> 점이 뒤로 이동
# e 뒤에 음수 -> 점이 앞으로 이동
print(1e5)
print(3.14e-3)
```

### 문자 리터럴
- 문자를 나열하여 입력한 것을 문자열(string)이라 하고, 이 일련의 문자열을 표기하는 것을 문자 리터럴(string literal)
- 작은 따옴표(') or 큰 따옴표(")로 묶어서 표기

```python
msg = 'hello, python!'
msg
```

```python
type(msg)
```

### 이스케이프 시퀀스
- 작은 따옴표는 문자열을 나타내는 특수 문자이기에, 출력하고 싶을 때 역슬래시(`\`)를 사용
- 줄바꿈(`\n`), 탭(`\t`), 작은 따옴표(`\'`), 큰 따옴표(`\"`)

```python
print('hello, \npython!')
print('hello,' + " python!") # 파이썬은 따옴표 구분 안함, 문자열 덧셈 가능
print('hello, '        'python!') # 공백을 사이에 두고 있으면 연속해서 입력한 것으로 간주
```

### raw 문자열 리터럴
- 이스케이프 시퀀스를 그대로 처리
- 문자열 앞에 r이나 R을 입력해서 사용
- 경로 등을 입력하는 경우에 사용

```python
print(r'C:\Source\kdt-python-2025')

# f-string 포매팅 : .format 하지 않아도 f'문자열 {변수} 문자열'로 만들 수 있음
s = '일'
print(f'나는 하고 싶은 {s}을 하면서 살고 싶다.')
```
