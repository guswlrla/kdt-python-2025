pi = 원둘레 / 지름 = 3.14159... /1

원둘레 = 2pi r (r은 반지름)

###중2 코딩 이야기###

1. 실험 공간: 정사각형 (네모)

가로: 2, 세로: 2

좌표: x∈[−1,1], y∈[−1,1]

면적: 2×2=4

✅ 2. 원의 조건
중심: (0, 0)

반지름: 1

원의 면적: pi * r*r = pi

비율 = 원면적/네모 면적 = pi/4
 
 (원에 있는 점의 수)/(전체 점의 수) = pi/4

```python
import random

def estimate_pi(n_points):
    inside_circle = 0
    for _ in range(n_points):
        x = random.uniform(-1, 1)
        y = random.uniform(-1, 1)
        if x**2 + y**2 <= 1:
            inside_circle += 1
    return 4 * inside_circle / n_points

print("근사한 π 값 =", estimate_pi(1000)) # 백만개, 천만개, 억개로 증가 시켜서 계산 
```

```python
# n_points = 1000 ~ 10000 까지 1000씩 증가
for n in range(1000, 100001, 1000):
    pi_est = estimate_pi(n)
    print(f"n_points={n:5d} → 근사 π = {pi_est}")
```

### read() / readline() / readlines() 와 비교

- file.read() > 파일 전체를 한번에 문자열로 읽음

- file.readlines() > 모든 줄을 리스트로 읽음

- file.readline() > 한 줄씩 읽음

- for line in file > 한 줄씩 자동 반복 

```python
# Student 클래스 정의
class Student:
    def __init__(self, name, age, grade):
        self.name = name
        self.age = age
        self.grade = grade

    def __str__(self):
        return f"이름: {self.name}, 나이: {self.age}, 학년: {self.grade}"

# 파일에서 데이터 읽기 및 Student 객체 생성
students = []

with open('students.txt', 'r', encoding='utf-8') as file:
    for line in file:
        # 각 줄의 데이터를 쉼표로 분리
        name, age, grade = line.strip().split(', ')
        # Student 객체 생성
        student = Student(name, int(age), grade)
        students.append(student)

# 학생 객체 출력
for student in students:
    print(student) # student는 객체 - __str__() 자동 호출
```

### AI·데이터 분석에서는 파일 예외 처리(file exception handling) 가 매우 중요한 이유

- 데이터 분석은 외부 파일을 다룸 : CSV/JSON/엑셀/로그/이미지 등 다양한 파일 포맷을 다루기 때문에 오류가 발생할 확률이 높음

- 값 누락·열 부족·잘못된 형식이 있으면 잘못된 학습/결과 발생

- 모델이 잘못된 데이터를 학습하면 품질 하락, 재학습 비용 증가

### 실무에서 가장 많이 발생하는 파일 문제 유형

- 파일 없음 : 지정된 데이터셋 파일/경로가 존재하지 않음

- 잘못된 구분자 : 콤마 대신 탭·세미콜론

- 컬럼 불일치 : 헤더 개수가 맞지 않음, 누락

- 결측치 : NaN, NULL, 빈 문자열

- 데이터 타입 오류 : 숫자 → 문자열 / 날짜 파싱 실패

```python
# Student 클래스 정의
class Student:
    def __init__(self, name, age, grade):
        self.name = name
        self.age = age
        self.grade = grade

    def __str__(self):
        return f"이름: {self.name}, 나이: {self.age}, 학년: {self.grade}"


students = []
filename = "students2.txt"

try:
    # 파일 열기 예외 처리
    with open(filename, 'r', encoding='utf-8') as file:
        for line_num, line in enumerate(file, start=1):
            try:
                # 공백 제거 및 필드 분리
                parts = line.strip().split(', ')
                if len(parts) != 3:
                    raise ValueError(f"필드 개수가 3개가 아님 → {parts}")

                name, age, grade = parts

                # 나이 정수 변환 예외 처리
                try:
                    age = int(age)
                except ValueError:
                    raise ValueError(f"나이가 정수가 아님 → '{age}'")

                # Student 객체 생성
                student = Student(name, age, grade)
                students.append(student)

            except Exception as e:
                print(f"[경고] {line_num}번째 줄에서 오류 발생: {e}")
                # 오류가 있어도 다음 줄 데이터 계속 처리

except FileNotFoundError:
    print(f"[오류] 파일을 찾을 수 없습니다: {filename}")

except Exception as e:
    print(f"[오류] 파일을 읽는 중 알 수 없는 에러: {e}")


# 정상적으로 읽힌 학생 객체 출력
print("\n=== 학생 정보 출력 ===")
for student in students:
    print(student)
```

### 13장 파일 실습 과제1
- data.csv 파일 읽어 클래스 객체로 만들기


```python
import csv
from datetime import datetime

# 1) 레코드를 담을 클래스 정의
class Record:
    def __init__(self, date_str, text):
        # date_str: "2021-01-01" 같은 문자열
        self.date_str = date_str
        self.text = text

        # 날짜를 datetime 객체로도 저장해두고 싶으면:
        try:
            self.date = datetime.strptime(date_str, "%Y-%m-%d").date()
        except ValueError:
            # 형식이 다르면 그대로 문자열만 사용
            self.date = None

    def __str__(self):
        # date가 파싱됐으면 예쁘게, 아니면 원래 문자열 출력
        if self.date is not None:
            return f"[{self.date}] {self.text}"
        else:
            return f"[{self.date_str}] {self.text}"


# 2) CSV 파일 읽어서 Record 객체 리스트에 저장
records = []

# data.csv 파일이 현재 파이썬 실행 파일(또는 노트북)과 같은 폴더에 있다고 가정
file_path = "data/data.csv"

with open(file_path, "r", encoding="utf-8") as f:
    reader = csv.reader(f)

    # 첫 줄이 헤더: date,text 인 경우 헤더 건너뛰기
    header = next(reader, None)
    # header가 실제로 'date','text'인지 확인해봐도 됨 (선택)

    for row in reader:
        if len(row) < 2:
            continue  # 열이 부족한 줄 건너뜀

        date_str, text = row[0], row[1]
        record = Record(date_str, text)
        records.append(record)

# 3) 클래스 객체들 화면에 출력
print("=== data.csv 레코드 출력 ===")
for rec in records:
    print(rec)
```

### 13장 파일 실습 과제2
- 2000.csv 파일 읽어 클래스 객체로 만들기

```python
import csv
from datetime import datetime

# 1) 레코드를 담을 클래스 정의
class Record:
    def __init__(self, date_str, text):
        # date_str: "2021-01-01" 같은 문자열
        self.date_str = date_str
        self.text = text

        # 날짜를 datetime 객체로도 저장해두고 싶으면:
        try:
            self.date = datetime.strptime(date_str, "%Y%m%d%H%M%S")
        except ValueError:
            # 형식이 다르면 그대로 문자열만 사용
            self.date = None

    def __str__(self):
        # date가 파싱됐으면 예쁘게, 아니면 원래 문자열 출력
        if self.date is not None:
            return f"[{self.date.strftime('%Y-%m-%d, %H:%M:%S')}] {self.text}"
        else:
            return f"[{self.date_str}] {self.text}"


# 2) CSV 파일 읽어서 Record 객체 리스트에 저장
records = []

# data.csv 파일이 현재 파이썬 실행 파일(또는 노트북)과 같은 폴더에 있다고 가정
file_path = "data/2000.csv"

with open(file_path, "r", encoding="utf-8") as f:
    reader = csv.reader(f)

    # 첫 줄이 헤더: date,text 인 경우 헤더 건너뛰기
    header = next(reader, None)
    # header가 실제로 'date','text'인지 확인해봐도 됨 (선택)

    for row in reader:
        if len(row) < 2:
            continue  # 열이 부족한 줄 건너뜀

        date_str, text = row[1], row[3]
        record = Record(date_str, text)
        records.append(record)

# 3) 클래스 객체들 화면에 출력
print("=== 2000.csv 레코드 출력 ===")
for rec in records:
    print(rec)
```
