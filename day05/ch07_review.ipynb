- 이터러블 : 반복을 할 수 있는 객체 (for문으로 하나씩 꺼낼 수 있는 객체)

```python
# 리스트 이터러블
x = ['john', 'geoge', 'judy', 'tom']

for i in range(len(x)):
    print(x[i])

for i, name in enumerate(x): # 인덱스와 값을 줌, 중요!!!
    print(i, name)

for i, name in enumerate(x, 10): # 카운트를 10부터 시작
    print(i, name)

for i in x:
    print(i)
```

```python
# 점수를 입력받아 총점과 평균 구하기 1

print('총점과 평균을 구해봅시다.')
number = int(input('학생 수：'))

score = [None] * number

for i in range(number):
    score[i] = int(input('{}번 학생의 점수：'.format(i + 1)))

total = 0
for i in range(number):
    total += score[i]

print('총점은 {}점입니다.'.format(total))
print('평균은 {}점입니다.'.format(total / number))
```

```python
# 점수를 입력받아 총점과 평균 구하기 2(오류)

print('총점과 평균을 구해봅시다.')
number = int(input('학생 수：'))

score = [None] * number

for i, point in enumerate(score):
    point = int(input('{}번 학생의 점수：'.format(i + 1))) # point는 복사본이기 때문에 값을 입력해도 저장이 안됨

total = 0
for i in range(number):
    total += score[i] 

print('총점은 {}점입니다.'.format(total))
print('평균은 {}점입니다.'.format(total / number))
```

```python
# 점수를 입력받아 총점과 평균 구하기 3(enumerate로 총점 구하기)

print('총점과 평균을 구해봅시다.')
number = int(input('학생 수：'))

score = [None] * number

for i in range(number):
    score[i] = int(input('{}번 학생의 점수：'.format(i + 1)))

total = 0
for i, point in enumerate(score):
    total += point                       # i의 값은 불필요

print('총점은 {}점입니다.'.format(total))
print('평균은 {}점입니다.'.format(total / number))
```

```python
# 리스트 행렬
x = [[1, 2, 3], [4, 5, 6]]
print(x)
```

```python
# 리스트 복사
x2 = x # 복사가 아니라, 참조변수 x2가 x를 가리킴
print(x2)

x3 = x.copy() # 얕은 복사
x3[0][0] = -1
print(x3)

import copy

x4 = copy.deepcopy(x) # 깊은 복사
x4[0][0] = 1
print(x4)
```

```python
lst1 = [1, 2, 3, 4, 5]
lst2 = lst1.copy() # 값이 복사됨, 
lst1[2] = 9
print(lst1)
print(lst2) # 뭐라는거야 ㅅㅂ
```

```python
# 리스트 내포 표기(list comprehension)
x = [n for n in range(1, 8)]
print(x)

x2 = [f'{n}{"+" if n % 2 == 1 else "-"}' for n in range(1, 8)] # f스트링 사용
print(x2)

x3 = [n for n in range(1, 8) if n % 2 == 0] # 조건을 더 붙일 수 있음
print(x3)

x4 = [i * 10 + j for i in range(1, 3) for j in range(1, 4)] # for문 중첩 가능
print(x4)

x5 = [[i * 10 + j for i in range(1, 3)] for j in range(1, 4)] # 2차원 배열
print(x5)

x6 = [[1 if i == j else 0 for i in range(1, 3)] for j in range(1, 4)]
print(x6)
```
