```python
class Member:
    def __init__(self, name, weight):
        self._name = name # ì–¸ë”ìŠ¤ì½”ì–´ í•˜ë‚˜(_)ëŠ” "ì™¸ë¶€ì—ì„œ ì§ì ‘ ì ‘ê·¼í•˜ì§€ ë§ì•„ ì£¼ì„¸ìš”"ë¼ëŠ” ê´€ë¡€ì  ì˜ë¯¸
        ## íŒŒì´ì¬ì´ ì ‘ê·¼ì„ ë§‰ì•„ì£¼ëŠ” ê¸°ëŠ¥ì€ ì—†ë‹¤.
        self._weight = weight

    # -------- getter í•¨ìˆ˜ --------
    def get_weight(self):
        print("âš™ getter í˜¸ì¶œ")
        return self._weight

    # -------- setter í•¨ìˆ˜ --------
    def set_weight(self, w):
        print("âš™ setter í˜¸ì¶œ")
        if w > 0:
            self._weight = w
        else:
            self._weight = 0

    # -------- deleter í•¨ìˆ˜ --------
    def del_weight(self):
        print("âš™ deleter í˜¸ì¶œ")
        del self._weight # _weight ë³€ìˆ˜ëŠ” ì˜ë„ê°€ private ìš©ë„ì´ë‚˜ ì œí•œì„ ë‘ì§€ ì•ŠëŠ”ë‹¤-ë‹¨ìˆœ ì°¸ê³ ìš©

    # weight í”„ë¡œí¼í‹° ìƒì„± > _weight ëŒ€ì‹ ì— weightë¥¼ ì‚¬ìš©í•œë‹¤ëŠ” ì˜ë¯¸
    weight = property(get_weight, set_weight, del_weight, doc="ì²´ì¤‘ í”„ë¡œí¼í‹°")
```

```python
class Member:
    def __init__(self, name, weight):
        self._name = name
        self._weight = weight

    @property
    def weight(self):
        print("âš™ getter í˜¸ì¶œ")
        return self._weight

    @weight.setter
    def weight(self, w):
        print("âš™ setter í˜¸ì¶œ")
        self._weight = w if w > 0 else 0

    @weight.deleter
    def weight(self):
        print("âš™ deleter í˜¸ì¶œ")
        del self._weight
```

íŒŒì´ì¬ì˜ í´ë˜ìŠ¤ ë©”ì†Œë“œ(classmethod) ì™€ ì •ì  ë©”ì†Œë“œ(staticmethod) ì˜ ì°¨ì´

```python
class Member:
    total_count = 0   # í´ë˜ìŠ¤ ë³€ìˆ˜ (ëª¨ë“  ê°ì²´ê°€ ê³µìœ )

    def __init__(self, name):
        self.name = name # ì¸ìŠ¤í„´ìŠ¤ ë³€ìˆ˜ - ê°ì²´ ë³€ìˆ˜
        Member.total_count += 1

    # ì¸ìŠ¤í„´ìŠ¤ ë©”ì†Œë“œ â†’ selfë¡œ ê°œë³„ ê°ì²´ì— ì ‘ê·¼
    def hello(self):
        print(f"ì•ˆë…•í•˜ì„¸ìš”, ì €ëŠ” {self.name}ì…ë‹ˆë‹¤.")

    # í´ë˜ìŠ¤ ë©”ì†Œë“œ â†’ clsë¡œ í´ë˜ìŠ¤ ìì²´ì— ì ‘ê·¼
    @classmethod # 1ë²ˆ
    def show_total(cls): # 2ë²ˆ: cls ì‚¬ìš©
        print(f"íšŒì› ìˆ˜: {cls.total_count}ëª…") # Member.total_countì´ë‘ ê°™ìŒ

    # ì •ì  ë©”ì†Œë“œ â†’ í´ë˜ìŠ¤/ê°ì²´ì™€ ë¬´ê´€, ìœ í‹¸ë¦¬í‹° í•¨ìˆ˜
    @staticmethod
    def is_valid_name(name):
        return len(name) >= 2
```

```python
print(f'{Member.total_count}')
```

```python
kim = Member("ê¹€ì¤€í˜¸")
lee = Member("ì´ìˆ˜ì—°")
print(f'{lee.total_count}')
kim.hello()            # ì¸ìŠ¤í„´ìŠ¤ ë©”ì†Œë“œ
Member.show_total()    # í´ë˜ìŠ¤ ë©”ì†Œë“œ
print(Member.is_valid_name("í™"))   # ì •ì  ë©”ì†Œë“œ
print(Member.is_valid_name("A"))    # ì •ì  ë©”ì†Œë“œ
```
