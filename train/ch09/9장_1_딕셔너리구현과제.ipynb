```python
# 실습 1: 1부터 100 사이의 난수를 20개 생성하여 리스트에 저장
import random

# 빈 리스트 생성
random_numbers = []

# 난수를 20개 생성하여 리스트에 추가
for _ in range(20):
    random_numbers.append(random.randint(1, 100))

# 저장된 리스트 출력
print(f"난수 생성 = {random_numbers}")
```

```python
# 실습 2: s1, s2, ..., s20 형태의 학번 리스트 생성

# 빈 리스트 생성
string_list = []

# s1 ~ s20 생성하여 리스트에 추가
for i in range(1, 21):
    string_list.append(f"s{i}")

# 저장된 리스트 출력
print(f"sno 리스트 = {string_list}")
```

```python
# 실습 3: 학번과 점수를 키-값으로 갖는 딕셔너리 생성

students = string_list
scores = random_numbers

# 딕셔너리 생성
score_dic = {students[i]: scores[i] for i in range(20)}

# 결과 출력
print(f"학번과 점수 딕셔너리 = {score_dic}")
```

```python
# 실습 4: 점수를 기준으로 내림차순 정렬한 튜플 리스트 생성

# 딕셔너리의 (학번, 점수) 쌍을 튜플로 정렬
# item() 함수 - 딕셔너리에 있는 각각의 키와 값들의 쌍을 가져옴(튜플 형식)
# lambda() 함수 - 이름이 없는 간단한 함수(익명 함수), lambda 매개변수: 반환값 형태
sorted_scores = sorted(score_dic.items(), key=lambda item: item[1], reverse=True) # item은 items 리스트의 요소
# [(s1, 100), (s2, 200), (s3, 300)] -> 튜플하나가 item, item[0] = s3 / item[1] = 300

# 결과 출력
print(f"점수로 정렬된 튜플 리스트 = {sorted_scores}")
```

```python
# 실습 5: 정렬된 리스트를 다시 딕셔너리로 변환

sorted_score_dic = dict(sorted_scores)

# 결과 출력
print(f"점수로 정렬된 딕셔너리 = {sorted_score_dic}")
```

```python
# 실습 6: 상위 5개의 학생만 추출하여 딕셔너리로 저장

top_5 = dict(sorted_scores[:5])

# 결과 출력
print(f"상위 5개 = {top_5}")
```

```python
# 실습 7: 딕셔너리 → 튜플 리스트 변환 (key, value 쌍으로)

score_list = list(score_dic.items())

# 결과 출력
print(f"리스트로 변환된 딕셔너리: {score_list}")
```

```python
# 실습 8: enumerate를 활용하여 딕셔너리를 새로운 딕셔너리로 변환

# enumerate(..., start=1) - 각 요소에 순번(i)를 붙임
# i: (key, value) 형식으로 반복문 돌기, 변수는 i와 (key, value)
transformed_score_dic = {i: (key, value) for i, (key, value) in enumerate(score_dic.items(), start=1)}

# 결과 출력
print(f"변환된 딕셔너리: {transformed_score_dic}")
```
