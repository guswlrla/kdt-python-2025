### ğŸ§© STEP 1. ë¦¬ìŠ¤íŠ¸ ì²˜ë¦¬ í•¨ìˆ˜ (ê¸°ì´ˆ + comprehension)

```python
# ğŸ“Œ ê³¼ì œ 1
# ì ìˆ˜ ë¦¬ìŠ¤íŠ¸ë¥¼ ë°›ì•„ 70ì  ì´ìƒì´ë©´ 'í•©ê²©', ê·¸ë ‡ì§€ ì•Šìœ¼ë©´ 'ë¶ˆí•©ê²©'ìœ¼ë¡œ ë¶„ë¥˜ëœ ë¬¸ìì—´ ë¦¬ìŠ¤íŠ¸ë¥¼ ë°˜í™˜í•˜ì„¸ìš”.

# ì…ë ¥ ì˜ˆì‹œ: [85, 67, 90, 45]
# ì¶œë ¥: ['í•©ê²©', 'ë¶ˆí•©ê²©', 'í•©ê²©', 'ë¶ˆí•©ê²©']

# map(function, iterable) - ë°˜ë³µ ê°€ëŠ¥í•œ ê°ì²´ì˜ ê° ìš”ì†Œì— í•¨ìˆ˜ë¥¼ ì ìš©
scores = list(map(int, input().split(' ')))
# print(scores)

def classify_scores(scores):
    return ['í•©ê²©' if s >= 70 else 'ë¶ˆí•©ê²©' for s in scores]

classify_scores(scores)
```

```python
# ğŸ“Œ ê³¼ì œ 2
# ì´ë¦„ ë¦¬ìŠ¤íŠ¸ë¥¼ ë°›ì•„ ê° ì´ë¦„ì— ìˆœë²ˆì„ ë¶™ì¸ ë¦¬ìŠ¤íŠ¸ë¥¼ ë°˜í™˜í•˜ì„¸ìš” (enumerate() ì‚¬ìš©).

# ì¶œë ¥ ì˜ˆì‹œ: ['1. í™ê¸¸ë™', '2. ê¹€ìœ ì‹ ', ...]

names = list(input().split(' '))

def numbered_names(names):
    return [f"{i+1}. {name}" for i, name in enumerate(names)]

numbered_names(names)
```

### ğŸ§© STEP 2. íŠœí”Œ & zip() ì²˜ë¦¬ í•¨ìˆ˜

```python
# ğŸ“Œ ê³¼ì œ 3
# ì´ë¦„, ì ìˆ˜ ë‘ ë¦¬ìŠ¤íŠ¸ë¥¼ ë°›ì•„ "í™ê¸¸ë™: 85ì "ê³¼ ê°™ì€ ë¬¸ìì—´ ë¦¬ìŠ¤íŠ¸ë¥¼ ë°˜í™˜í•˜ì„¸ìš” (zip() ì‚¬ìš©).

names = list(input().split(' '))
scores = list(map(int, input().split(' ')))

def combine_name_score(names, scores):
    return [f"{name}: {score}ì " for name, score in zip(names, scores)]

combine_name_score(names, scores)
```

```python
# ğŸ“Œ ê³¼ì œ 4
# (ì´ë¦„, ì ìˆ˜) íŠœí”Œ ë¦¬ìŠ¤íŠ¸ì—ì„œ 80ì  ì´ìƒì¸ ì‚¬ëŒë§Œ í•„í„°ë§ëœ íŠœí”Œ ë¦¬ìŠ¤íŠ¸ë¥¼ ë°˜í™˜í•˜ì„¸ìš”.

names = list(input().split(' ')) # ['í™ê¸¸ë™', 'ê¹€ìœ ì‹ ', 'ê°•ê°ì°¬']
scores = list(map(int, input().split(' '))) # [85, 70, 60]

# ë¦¬í„°ëŸ´ - ì½”ë“œì— ì§ì ‘ ì“°ëŠ” ê°’ ([], () ë“±ìœ¼ë¡œ ë°”ë¡œ ì“°ëŠ” ê°’ ìì²´)
# ì´í„°ëŸ¬ë¸” - ë°˜ë³µ ê°€ëŠ¥í•œ ê°ì²´ (list, tuple, str, zip, map, range ë“±)
# ë¦¬í„°ëŸ´ ë°©ì‹ìœ¼ë¡œ zip()ì„ í•˜ë©´ zip ê°ì²´ ìì²´ê°€ ë¦¬ìŠ¤íŠ¸ ì•ˆì— ë“¤ì–´ê°, ì–¸íŒ¨í‚¹ x
data = list(zip(names, scores))
data1 = [zip(names, scores)]
print(data1)

def filter_high_scores(data):
    return [(name, score) for name, score in data if score >= 80]

filter_high_scores(data)
```

### ğŸ§© STEP 3. ë”•ì…”ë„ˆë¦¬ ìƒì„±/ë³€í™˜ í•¨ìˆ˜

```python
# ğŸ“Œ ê³¼ì œ 5
# ì´ë¦„ ë¦¬ìŠ¤íŠ¸ì™€ ì ìˆ˜ ë¦¬ìŠ¤íŠ¸ë¥¼ ë°›ì•„ ë”•ì…”ë„ˆë¦¬ë¡œ ë³€í™˜í•˜ëŠ” í•¨ìˆ˜ (zip() ì‚¬ìš©)

# ì…ë ¥: ["í™ê¸¸ë™", "ê¹€ìœ ì‹ "], [85, 90]
# ì¶œë ¥: {"í™ê¸¸ë™": 85, "ê¹€ìœ ì‹ ": 90}

names = list(input().split(' '))
scores = list(map(int, input().split(' ')))

def to_score_dict(names, scores):
    return {name: score for name, score in zip(names, scores)}

to_score_dict(names, scores)
```

```python
# ğŸ“Œ ê³¼ì œ 6
# í•™ìƒ ì ìˆ˜ ë”•ì…”ë„ˆë¦¬ë¥¼ ë°›ì•„ í•©ê²©ì ì´ë¦„ë§Œ ë¦¬ìŠ¤íŠ¸ë¡œ ë°˜í™˜í•˜ì„¸ìš” (cutoff ì§€ì • ê°€ëŠ¥)

names = list(input().split(' '))
scores = list(map(int, input().split(' ')))
# ì—¬ê¸°ë„ ë§ˆì°¬ê°€ì§€ë¡œ {}ë¡œ ê°ì‹¸ë©´ zip()ì˜ ê°ì²´ê°€ ë“¤ì–´ê°
score_dict = dict(zip(names, scores))
print(score_dict)

def passed_students(score_dict, cutoff=70):
    # item() í•¨ìˆ˜ - ë”•ì…”ë„ˆë¦¬ì— ìˆëŠ” ê°ê°ì˜ í‚¤ì™€ ê°’ë“¤ì˜ ìŒì„ ê°€ì ¸ì˜´(íŠœí”Œ í˜•ì‹)
    return [name for name, score in score_dict.items() if score >= cutoff]

passed_students(score_dict)
```

### ğŸ’  STEP 6: í™•ì¥ ê³¼ì œ (ì‹¤ì „í˜•)

```python
# ğŸ“Œ ê³¼ì œ 10: ë°ì´í„° ì „ì²˜ë¦¬ â†’ ë¶„ì„ê¹Œì§€ ì „ ê³¼ì • í•¨ìˆ˜ ë¶„ë¦¬ ì‹¤ìŠµ
# ì•„ë˜ì˜ ë¦¬ìŠ¤íŠ¸ë¥¼ ê¸°ë°˜ìœ¼ë¡œ ë‹¤ìŒ ì‘ì—…ì„ ê°ê°ì˜ í•¨ìˆ˜ë¡œ ë¶„ë¦¬í•˜ì—¬ êµ¬í˜„í•´ë³´ì„¸ìš”.

names = ["í™ê¸¸ë™", " ê¹€ìœ ì‹  ", "ì´ìˆœì‹ ", "ê°• ê° ì°¬"]
scores = [85, 70, 60, 90]
subjects = ["ìˆ˜í•™", "ì˜ì–´", "ê³¼í•™", "êµ­ì–´"]

# í•¨ìˆ˜ ë¶„ë¦¬ ëª©í‘œ
# clean_names() â†’ ì´ë¦„ì—ì„œ ê³µë°± ì œê±° ë° ì •ë¦¬
def clean_names(name_list):
    return [name.replace(' ', '') for name in name_list]

clean_names_lst = clean_names(names)

# to_dict() â†’ ì´ë¦„ê³¼ ì ìˆ˜ ë”•ì…”ë„ˆë¦¬ë¡œ ë³€í™˜
def to_dict(nameVal, scoreVal):
    return {name: score for name, score in zip(nameVal, scoreVal)}

to_dict(clean_names_lst, scores)

# filter_pass() â†’ 70ì  ì´ìƒì¸ í•™ìƒë§Œ í•„í„°ë§
def filter_pass(scores):
    # filter(function, iterable) - ì¡°ê±´ì— ë§ëŠ” ìš”ì†Œë§Œ ì„ íƒí•´ì„œ Trueì¸ ìš”ì†Œë§Œ ë‚¨ê¹€
    filterScore = filter(lambda i: i >= 70, scores)
    return list(filterScore)

filter_pass(scores)

# report() â†’ "1. í™ê¸¸ë™ - ìˆ˜í•™: 85ì " í˜•íƒœë¡œ ì¶œë ¥
def report(names, subjects, scores):
    # enumerate() - ìˆœì„œ ë¶™ì¼ë ¤ê³  ì‚¬ìš©
    # zip() - ('í™ê¸¸ë™', 'ìˆ˜í•™', 85)ë¡œ ë¬¶ìŒ
    # enumerateê°€ ë°˜í™˜í•˜ëŠ”ê²Œ (ì¸ë±ìŠ¤, zipìœ¼ë¡œ ë¬¶ì€ ìš”ì†Œ íŠœí”Œ) - (1, ('í™ê¸¸ë™', 'ìˆ˜í•™', 85))
    # iì™€ (name, sub, score)ë¼ëŠ” íŠœí”Œ ìì²´ë¥¼ í’€ì–´ì„œ ê°ê° ë³€ìˆ˜ë¡œ ë‚˜ëˆ”(ì–¸íŒ¨í‚¹)
    return [f"{i}. {name} - {sub}: {score}ì " for i, (name, sub, score) in enumerate(zip(names, subjects, scores), start=1)]

report(clean_names_lst, subjects, scores)
```
